```python
'''
Word Embedding
     
     Word embeddings capture semantic relationships between words, allowing models to understand and 
represent words in a continuous vector space where similar words are close to each other. This semantic representation 
enables more nuanced understanding of language.

Frequency-based embedding 
        Term Frequency-Inverse Document Frequency (TF-IDF)
        Co-occurrence Matrix

Prediction-based embeddings
        Word2Vec (Skip-gram and Continuous Bag of Words)
        FastText
        GloVe (Global Vectors for Word Representation)
        
'''
```

```python
'''
Word2vec captures semantic meaning in sentence like 'happy' &  'joy' were seperate termaccording to TF-IDF but 
word2vec captures its symantic meaning.

Low dimension vectors compared to TF-IDF, So we have fast computing

Word2vec gives us dense vectors so we dont have overfitting problem

'''
```

CBOW (Continuous Bag of Words)

```python
'''
CBOW uses sliding window concept say


'watch youtube for data science' is the sliding window is of 3 (3,5,7,...)

                        --------------
                ___X___ | target = Y | ___X___ (for window length 3)
                        --------------
         ---------
  watch | youtube | for
         ---------
         
           -----
  youtube | for | data
           -----      
         
       ------
  for | data | science
       ------      
         
  
  X (known value)              Y (predict)
  
  watch, for                      youtube
  
  youtube, data                     for
  
  for, science                     data

  
  Here, the words will be one hot encoded like
  
  watch       1 0 0 0 0
  youtube     0 1 0 0 0 
  for         0 0 1 0 0 
  data        0 0 0 1 0
  science     0 0 0 0 1
  
  This will be fed to a NN.
  
=====================================================================================================================
                                                    ------------
  If sliding window is of length 5 ___X___ ___X___ | target = Y | ___X___ ___X___ 
                                                    ------------
  
                                                            ------------
  If sliding window is of length 7 ___X___ ___X___ ___X___ | target = Y | ___X___ ___X___ ___X___
                                                            ------------
                                                            
=====================================================================================================================

Structure of CBOW in NN is

If word length is n
i.e,  'watch youtube for data science' has n = 5
       w = length of sliding window (3 in above example)

  - [1]

  - [2]       - [1]         - [1]  ----
  
  - [3]       - [2]         - [2]  ----           
  
  - [4]       - [3]           .
                              .
   .            .           - [n]  ----
   .            .
   .          - [w]
   
 - [2n] 



  1 I/P layer with 2n inputs, hidden layer with w neurons and n O/P layers for O/P prediction. The connection will 
  be densely connected layers

  - [ 1]           

  - [ 2]       
  
  - [ 3]                     [1] --->   
                             
  - [ 4]      - [1] -        [2] ---> 
                              
  - [ 5]      - [2] -        [3] ---> 
                          
  - [ 6]      - [3] -        [4] --->  
   
  - [ 7]                     [5] ---> 
  
  - [ 8]
  
  - [ 9]
  
  - [10]
  
=====================================================================================================================
  
  X (known value)              Y (predict)
  
  watch, for                      youtube
  
  youtube, data                     for
  
  for, science                     data

  
  Here, the words will be one hot encoded like
  
  watch       1 0 0 0 0
  youtube     0 1 0 0 0 
  for         0 0 1 0 0 
  data        0 0 0 1 0
  science     0 0 0 0 1
  
  First input will be watch, for and NN will predict youtube
  
  
X                                         Y

1   - [ 1]            
0   - [ 2]       
0   - [ 3]                     [1] ---> watch                       
0   - [ 4]      - [1] -        [2] ---> youtube                             
0.  - [ 5]      - [2] -        [3] ---> for                        
0   - [ 6]      - [3] -        [4] ---> data  
0   - [ 7]               ^     [5] ---> science
1   - [ 8]               |
0   - [ 9]               |
0.  - [10]            softmax
                    activation
                     function
  
'''
```

Skip-Gram

```python
'''
Skip-Gram is complete opposite of CBOW.

'watch youtube for data science' is the sliding window is of 3 (3,5,7,...)

                ------------           ------------
               | target = Y | ___X___ | target = Y |      (for window length 3)
                ------------           ------------ 

     -------           -----
    | watch | youtube | for |  
     -------           -----

     ---------       ------
    | youtube | for | data |
     ---------       ------
     
     -----        --------
    | for | data | science |
     -----        --------
     
 X (known value)              Y (predict)
 
 youtube                    watch, for
 
 for                        youtube, data
 
 data                       for, science
  
  Here, the words will be one hot encoded like
  
  watch       1 0 0 0 0
  youtube     0 1 0 0 0 
  for         0 0 1 0 0 
  data        0 0 0 1 0
  science     0 0 0 0 1
  
  This will be fed to a NN.
  
====================================================================================================================================
                                    
                                    ------------   ------------           ------------   ------------
  If sliding window is of length 5 | target = Y | | target = Y | ___X___ | target = Y | | target = Y |
                                    ------------   ------------           ------------   ------------
  
                                                
                                    ------------   ------------   ------------           ------------   ------------   ------------
  If sliding window is of length 7 | target = Y | | target = Y | | target = Y | ___X___ | target = Y | | target = Y | | target = Y |
                                    ------------   ------------   ------------           ------------   ------------   ------------
                                                            
====================================================================================================================================

Structure of Skip-Gram

If word length is n
i.e,  'watch youtube for data science' has n = 5
       w = length of sliding window (3 in above example)

                              
                              [1] --->

  -[1]-                       [2] --->
  
  -[2]-       - [1]-          [3]  --->
  
  -[3]-       - [2] -          .
  
    .            .             .
    
    .            .             .
    
    .          -[w] -          .
  
  -[n]-                        .
                             
                               [2n] --->


  1 I/P layer with n inputs, hidden layer with w neurons and 2n O/P layers for O/P prediction. The connection will 
  be densely connected layers
  
                              [1] --->
  
                              [2] --->
  
                              [3]  --->

  -[1]-                       [4]  --->
  
  -[2]-       - [1]-          [5]  --->
  
  -[3]-       - [2]-          [6] --->
  
  -[4]-       - [3]-          [7] --->
    
  -[5]-                       [8] --->
  
                              [9] --->
                              
                              [10] --->
                              
        
 X (known value)              Y (predict)
 
 youtube                    watch, for
 
 for                        youtube, data
 
 data                       for, science
 

  Here, the words will be one hot encoded like
  
  watch       1 0 0 0 0
  youtube     0 1 0 0 0 
  for         0 0 1 0 0 
  data        0 0 0 1 0
  science     0 0 0 0 1
  
  First input will be watch, for and NN will predict youtube
  
                              [1] ---> 1
                              [2] ---> 0
                              [3] ---> 0
0 -[1]-                       [4] ---> 0
1 -[2]-       - [1]-          [5] ---> 0.
0 -[3]-       - [2]-          [6] ---> 0
0 -[4]-       - [3]-          [7] ---> 0
0 -[5]-                       [8] ---> 1
                              [9] ---> 0                       
                              [10] ---> 0.

''' 
```

```python
'''
Points to remember

For small data use CBOW, for large data use Skip-gram

Ways to improve Word2Vec efficiency:

        increasing training data
        
        increse dimension of hidden layer from w to higher dimension w+1,w+2,w+3,...
        
        incresing window size in itself i.e. w=5 is better than w=3, w=7 is better than w=5

'''
```

```python
#training a model in Word2Vec
#https://www.kaggle.com/datasets/khulasasndh/game-of-thrones-books
#Training word2vec using game of thrones books dataset

import numpy as np
import pandas as pd
```

```python
#import gensim
import os
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
file_list=['001ssb.txt','002ssb.txt','003ssb.txt','004ssb.txt','005ssb.txt']

story = []

for filename in file_list:
    filename=open(filename)
    corpus = filename.read()
    raw_sent = sent_tokenize(corpus)
    
    for sent in raw_sent:
        story.append(simple_preprocess(sent))
        
    print(story)
```

```python
story
```

```python
len(story)
```

```python
model = gensim.models.Word2Vec(
        window=10,#window size = 10, 10 words on left and right for predicting 11th word from left
        min_count=2, #we will only take sentence having atleast 2 words
)#press shift + tab to see more parameters
```

```python
model.build_vocab(story)#we have lot of words in corpus and here we will extract unique words using build_vocab() from story
```

```python
model.train(story,total_examples=model.corpus_count,epochs=model.epochs)
#this will train our deep learning model 
#passing entire corpus ---> story
#total_examples means total sentences in corpus given by model.corpus_count
#epochs=model.epochs=5 epoch(default value)
```
